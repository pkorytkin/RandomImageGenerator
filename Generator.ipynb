```python
import numpy as np
from PIL import Image,ImageOps
from matplotlib import cm
```

```python
SizeX=64
SizeY=64
```

```python
ImageArray=np.zeros((SizeX,SizeY,3))
print(ImageArray.shape)
for x in range(0,SizeX):
    for y in range(0,SizeY):
        for channel in range(3):
            ImageArray[x][y][channel]=np.random.randint(0,255)
#print(ImageArray)
arrayInt8=ImageArray.astype(np.uint8)
im = Image.fromarray(arrayInt8)
#im.show()
display(im)
im.save("Ready.png")
im2=ImageOps.grayscale(im)
im2.save("Ready2.png")
```
